```python
# create a column  age_group:
labels=['young','adult','mid-age','senior']
df['age_group']= pd.qcut(df['age'], q=4,labels=labels)
        
```

```python
# create a column pruchase_frequency:
frequency_mapping={
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly':30,
    'Bi-Weekly':14,
    'Quarterly':90,
    'Anually': 365,
    'Every 3 months' : 90
}
df['purchase_frequency']=df['frequency_of_purchases'].map(frequency_mapping)
```
